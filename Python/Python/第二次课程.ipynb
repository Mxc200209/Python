## 计算收益率
- 一般收益率使用收盘价格计算

- $R_t = (p_t - p_{t-1}) / p_{t-1}$

- Log_return: $r_t = log(p_t) - log(p_{t-1})$

```python
index = data['2000-01':'2022-07'].copy()
index
```

```python
index['Close'] = pd.to_numeric(index['Close'])
index['Preclose'] = pd.to_numeric(index['Preclose'])
index
```

```python
# 计算上证指数的收益率 效率更高 向量计算
index['Raw_return'] = index['Close'] / index['Preclose'] - 1
index['Log_return'] = np.log(index['Close']) - np.log(index['Preclose'])
index['Stupid_return'] = 0
index
```

```python
for i in range(0,5470):
    index.iloc[i,9] = index.iloc[i,4] / index.iloc[i,0] -1
    # tab自动对齐 判断优先级
index
```

# 计算月的收益率 Calculate monthly return

- 大部分交易策略和学术都是使用月频度

- 策略一般高频策略，低频月度

- 学术基本以月度为主

# resample函数

```python
Month_data = index.resample('M')['Log_return'].sum().to_frame()  #M-月度 Q-季度 Y-年度
Month_data['Raw_return'] = np.exp(Month_data['Log_return'])-1
Month_data
```

```python
Month_data2 = index.resample('M')['Close'].last().to_frame()
Month_data2['Preclose'] = Month_data2['Close'].shift(1)
Month_data2['Log_return'] = np.log(Month_data2['Close']) - np.log(Month_data2['Preclose'])
Month_data2
```

```python
# 季度
Quarter_data = index.resample('Q')['Log_return'].sum().to_frame()  
Quarter_data['Raw_return'] = np.exp(Quarter_data['Log_return'])-1
Quarter_data
```

```python
Quarter_data2 = index.resample('Q')['Close'].last().to_frame()
Quarter_data2['Preclose'] = Quarter_data2['Close'].shift(1)
Quarter_data2['Log_return'] = np.log(Quarter_data2['Close']) - np.log(Quarter_data2['Preclose'])
Quarter_data2
```
