```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #画图模块
from datetime import datetime #时间模块
from IPython.core.interactiveshell import InteractiveShell

#输出矢量图 渲染矢量图 魔法函数（Magic Functions）内嵌绘图
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

#显示每一个运行结果
InteractiveShell.ast_node_interactivity = 'all'

#输出或者打印的时候，不限制列或者行
pd.set_option('display.max_columns',None)
#pd.set_option('display.max_rows',None)
```

```python
data = pd.read_csv('C:/Users/hp/Desktop/Python/Python-3/000001.csv')
data['Day'] = pd.to_datetime(data['Day'], format = '%Y-%m-%d')
data.set_index('Day', inplace = True)
data.sort_values(by = ['Day'], axis=0, ascending=True)
data
```

```python
index = data['1995-01':'2022-07'].copy()
index['Close'] = pd.to_numeric(index['Close'])
index['Preclose'] = pd.to_numeric(index['Preclose'])
index['Raw_return'] = index['Close'] / index['Preclose'] - 1
index['Log_return'] = np.log(index['Close']) - np.log(index['Preclose'])
index
```

```python
Month_data = index.resample('M')['Log_return'].sum().to_frame()  
Month_data['Raw_return'] = np.exp(Month_data['Log_return'])-1
Month_data
```

```python
Quarter_data = index.resample('Q')['Log_return'].sum().to_frame()  
Quarter_data['Raw_return'] = np.exp(Quarter_data['Log_return'])-1
Quarter_data
```

## 中国上证指数000001年度回报率数据

```python
Year_data = index.resample('Y')['Log_return'].sum().to_frame()  
Year_data['Raw_return'] = np.exp(Year_data['Log_return'])-1
Year_data
```

```python
Year_data2 = index.resample('Y')['Close'].last().to_frame()
Year_data2
```

## 数据可视化

```python
%%time
fig = plt.figure(figsize = (15,5))
plt.plot('Raw_return', #画图的变量名
         '.-c', #线的类型   
         label = 'Return', #线的类别 主要是在多线条画图的时候，起到区别的作用；对单线条没有影响
         linewidth = 1, #线的粗细
         data = Month_data['1995-01-01':'2022-07-31'])
         
plt.title("China's Stock Market") #标题
plt.xlabel('Month') #横坐标标题
plt.ylabel('Return') #纵坐标标题

year_freq = pd.date_range(start = '1995', end = '2022', freq = 'y')
c = plt.xticks(year_freq, year_freq.year, rotation = 90, fontsize = 10)

fig.savefig('month_return.pdf', bbox_inches = 'tight') #保存图片全部内容
```

```python
%%time
fig = plt.figure(figsize = (15,5))
plt.plot('Raw_return', 
         '*-m',    
         label = 'Return', 
         linewidth = 2, 
         data = Year_data['1995':'2021'])
         
plt.title("China's Stock Market") 
plt.xlabel('Year') 
plt.ylabel('Return') 

year_freq = pd.date_range(start = '1995', end = '2021', freq = 'y')
c = plt.xticks(year_freq, year_freq.year, rotation = 90, fontsize = 10)

fig.savefig('month_return.pdf', bbox_inches = 'tight')
```
